```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta, timezone
```

```python
JST = timezone(timedelta(hours=+9), 'JST')
```

```python
df = {}
date_offset = 5

for i in range(10):
    df[i] = pd.read_csv('./logger/usbrh_2022-01-{:02d}.dat'.format(i+date_offset), delim_whitespace=True)
    pass
```

```python
dtime_array = np.array([])
temp_log = np.array([])

for i in range(10):
    dtime_array = np.append(dtime_array, np.array([datetime.strptime(j, '%Y-%m-%d_%H:%M:%S-%Z') for j in df[i]['#Date']]))
    temp_log = np.append(temp_log, np.array(df[i]['Temp[deg]']))
    pass
```

```python
plt.plot(temp_log, 'bo')
```

```python
fig = plt.figure(figsize=(20,9))
plt.plot(dtime_array, temp_log, 'bo', markersize=2)
plt.ylim([10,50])
```

```python
fig,axarr = plt.subplots(1,2,figsize=(20,9))
ax = axarr.ravel()

ax[0].plot(dtime_array, temp_log, 'bo')
ax[0].set_xlim([dtime_array[-500], dtime_array[-1]])
ax[0].set_ylim([10,50])

ax[1].plot(dtime_array, temp_log, 'bo')
ax[1].set_xlim([dtime_array[-400], dtime_array[-280]])
ax[1].set_ylim([47,48])
```
